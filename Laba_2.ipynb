# 5. Избавимся от пустых значений, заменив их на медианные значения для количественных признаков и на последнее встречающееся значения с помощью метода “ffill” в аргументах функции fillna для категориальных:

```python
for namecol in tb.columns:
    if(tb[namecol].dtype == np.float64 or tb[namecol].dtype == np.int64):
        tb = tb.fillna(value = {namecol: tb[namecol].median()})
tb = tb.fillna(method = 'ffill')
tb = tb.fillna(value = {'Cabin':'C85'}, limit = 1)
tb
```

# 9. Правда ли, что люди моложе 30 лет выживали чаще, чем люди старше 60 лет?
## Доли выживших в обеих группах:

```python
under30 = tb[(tb["Age"] < 30) & (tb["Survived"] == 1)]["PassengerId"].count() / tb[(tb["Age"] < 30)]["PassengerId"].count()
over60 = tb[(tb["Age"] > 60) & (tb["Survived"] == 1)]["PassengerId"].count() / tb[(tb["Age"] > 60)]["PassengerId"].count()
print('Доля выживших людей младше 30 -',round(under30,2),', доля выживших людей старше 60 -',round(over60,2))
if (under30 > over60):
    print("Люди моложе 30 лет выживали чаще чем люди старше 60 лет.")
elif (under30 < over60):
    print("Люди старше 60 лет выживали чаще чем люди моложе 30 лет.")
else: print("Доля выживших младше 30 равна доле выживших старше 60.")
```

# 10. Правда ли, что женщины выживали чаще мужчин?
## Доли выживших в обеих группах:

```python
men = tb[(tb["Sex"] == "male") & (tb["Survived"] == 1)]["PassengerId"].count() / tb[(tb["Sex"] == "male")]["PassengerId"].count()
women = tb[(tb["Sex"] == "female") & (tb["Survived"] == 1)]["PassengerId"].count() / tb[(tb["Sex"] == "female")]["PassengerId"].count()
print('Доля выживших среди мужчин -',round(men,2),', Доля выживших среди женщин -',round(women,2))
if (men > women):
    print("Мужчины выживали чаще чем женщины.")
elif (men < women):
    print("Женщины выживали чаще чем Мужчины.")
else: print("Доля выживших среди мужчин равна доле выживших среди женщин.")
```

# 19. Как факт выживания зависит от возраста пассажира? Проверим(графически) предположение, что молодые чаще выживали. Пусть, условно, молодые - младше 30 лет, пожилые – старше 60 лет.

```python
tb.insert(1, "interval_age", 0)
tb["interval_age"] = np.where((tb.Age <= 30), "0-30", tb.interval_age)
tb["interval_age"] = np.where(((tb.Age < 60) & (tb.Age > 30)), "30-60", tb.interval_age)
tb["interval_age"] = np.where((tb.Age >= 60), "60-...", tb.interval_age)
sns.countplot(x = tb["interval_age"], hue = tb["interval_age"], edgecolor='black',)
print(' ')
```
