## Модуль B. Классификация
### ОТБОР ПРИЗНАКОВ
#### Необходимо определить, какие признаки имеют наибольшее влияние на классификацию объектов по Строительной категории (building_class_category) и оставить только их для обучения. Могут остаться и все исходные признаки. Необходимо обосновать выбор признаков, оставленных для обучения.

```python
bsm_B = pd.read_csv('bsm_data4.csv')
```

```python
bsm_B = bsm_B.drop(['Unnamed: 0',], axis=1)
```

```python
#list(bsm_B.columns)
pd.set_option('display.max_columns', None)
bsm_B.head()
```

```python
label = LabelEncoder()
for k in range(bsm_B.shape[1]):
    if(bsm_B[bsm_B.columns[k]].dtype == 'O'):
        bsm_B[bsm_B.columns[k]]=label.fit_transform(bsm_B[bsm_B.columns[k]])
bsm_B.head()
```

Проведём отбор признаков, оказывающих наибольшее влияние.

```python
bsm_B_test = bsm_B.drop("building_class_category", axis = 1)

x = bsm_B_test.copy() # копируем наш датафрейм в переменную х
y = bsm_B['building_class_category'] # создаем y - целевую переменную

from sklearn.model_selection import train_test_split # подгружаем функцию, которая разбивает данные случ. образом

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.33)#, random_state = 1)
# разбиение данных на обучающую и проверочную выборки

y_train
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_cls = RandomForestClassifier(n_estimators=300)

rf_cls.fit(x_train, y_train)
```

```python
#Исследование столбцов на их значимость
pd.set_option('display.max_rows', 20)

coef_rf = pd.DataFrame(rf_cls.feature_importances_, x_train.columns, columns=["coef_rf"]).sort_values(by = "coef_rf", ascending = False)

coef_rf.head(10)
```

```python
#Выделение бесполезных столбцов (влияение меньше 0.01)
names_str = coef_rf.index
delete_str = []

for i in range(names_str.shape[0]):
    if coef_rf['coef_rf'][i] >= 0.01:
        delete_str.append(names_str[i])
        
coef_rf = coef_rf.drop(delete_str, axis=0)
coef_rf.shape[0]
```

```python
#удаление бесполезных столбцов
names_str = coef_rf.index
names_col = bsm_B.columns
delete_col = []

for i in range(bsm_B.shape[1]):
    for j in range(names_str.shape[0]):
        if bsm_B.columns[i] == coef_rf.index[j]:
            delete_col.append(names_col[i])

bsm_B = bsm_B.drop(delete_col, axis=1)
bsm_B.shape[1]
```

```python
#сохранение данных в файл, чтобы в случае чего не приходилось считать по несколько раз одно и то же.
bsm_B.to_csv('bsm_dataB.csv')
```
